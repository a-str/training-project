## Предобработка данных

### Обработка пропусков

```python
# удаление явных дубликатов сторок
df = df.drop_duplicates().reset_index(drop=True)
df.info()
```

```python
# подсчёт явных пропусков
df.isna().sum()
```

Избавимся от данных с пропуском в `locality_name` - это ключевой параметр и таких квартир немного.
Пропуски `ceiling_height` заменим медианой для данного населенного пункта и домов с аналогичной этажностью.  Пропуски в `ponds_nearest`, `ponds_around3000`, `parks_nearest`, `parks_around3000`, `cityCenters_nearest`, `airports_nearest`, `balcony,` `kitchen_area`, `is_apartment`, `living_area`, `floors_total` оставим, т.к. это позволит сохранить инфорацию других полей объявлений и заменим на медианные значения для населенного пункта.

```python
#удаление строк с пропусками locality_name
df.dropna(subset=['locality_name'], inplace=True)
```

```python
# проверим уникальные названия населенных пунктов
df['locality_name'].unique()
```

```python
#заменим в названиях "ё" на "е"
df['locality_name'] = df['locality_name'].str.replace('посёлок', 'поселок')
```

```python
# замена пропусков ceiling_height медианой для квартир данного населенного пункта
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby(['locality_name'])['ceiling_height'].transform('median'))
df['ceiling_height'].value_counts()
```

```python
# заполним пропуски площади кухни для аппартаментов на 0
df.loc[(df['kitchen_area'].isna())&(df['is_apartment'] == True), 'kitchen_area'] = 0
```

```python
# проверка количества комнат в квартирах
df['rooms'].sort_values().unique()
```

```python
# замена числа комнат 0 на 1
df.loc[df['rooms'] == 0, 'rooms'] = 1
df['rooms'].value_counts()
```

```python
# замена оставшихся пропусков медианой для квартир данного населенного пункта
list_col = ['parks_nearest', 'ponds_nearest', 'airports_nearest', 'cityCenters_nearest']
for col in list_col:
    df[col] = df[col].fillna(df.groupby(['locality_name'])[col].transform('median'))
```

```python
# заменим оставшиеся пропуски полей list_col медианой по всем данным
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
for col in list_col:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isna().sum()
```

```python
# заменим пропуски количества балконов на 0
df['balcony'] = df['balcony'].fillna(0)
```

```python
# заменим пропуски is_apartment на False, как не относящиеся к аппартаментам
df['is_apartment'] = df['is_apartment'].fillna(False)
```

```python
# заменим пропуски дней экспозиции объявления days_exposition на 0
df['days_exposition'] = df['days_exposition'].fillna(0)
```

```python
# заполним пропуски жилой площади на разницу общей площади total_area и площади кухни kitchen_area
df.loc[df['living_area'].isna(), 'living_area'] = df.loc[df['living_area'].isna(), 'total_area'] - df.loc[df['living_area'].isna(), 'kitchen_area']
```

```python
# проверка отсутствия пропусков данных
df.isna().sum()
```

### Замена типа данных

```python
# приведение is_apartment к типу bool не требуется
df['is_apartment'].unique()
```

```python
# проверка неявных пропусков balcony
df['balcony'].unique()
```

```python
# приведение balcony к типу int64
df.loc[df['balcony'], 'balcony'] = df.loc[df['balcony'], 'balcony'].astype('int64')
df['balcony'].value_counts()
```

```python
#  приведение days_exposition к типу int64
df['days_exposition'] = df[df['days_exposition'].notna()]['days_exposition'].astype('int64')
df['days_exposition'].value_counts()
```

```python
# приведем к формату данных поле first_day_exposition
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

```python
# приведение данных формата float64 в полях parks_around3000, parks_nearest, ponds_around3000, ponds_nearest к формату int64
list_int64 = ['parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest', 'floors_total', 'cityCenters_nearest', 'airports_nearest', 'last_price']
for col in list_int64:
    try:
        df[col] = df[df[col].notna()][col].astype('int64')
    except:
        print('Ошибка конверсии формата')

df.info()
```

**Вывод**

Все данные приведены к целевому типу, обработаны все явные пропуски. Возможная причина появления пропусков - ошибки заполения данных объявления, а также пропуски в выгрузке данных из информационных систем - таких как  по расстоянию до ближайшего аэропорта или озера, при отсутствии последних в населенном пункте. Пропуски kitchen_area частично обусловлены отстутствием кухни в аппартаментах. Неявных пропусков не выявлено.

## Расчёты и добавление результатов в таблицу

Расчет цены квадратного метра:

```python
df['qm_price'] = (df['last_price'] / df['total_area']).astype('int64')
```

Расчет дня недели, месяца и года публикации:

```python
df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
```

Расчет этажа квартиры (первый, последний, другой):

```python
# функция категоризации этажа квартиры
def floor_categorization(row_floor):
    if row_floor['floor'] == 1:
        return 'первый'
    if row_floor['floor'] >= row_floor['floors_total']:
        return 'последний'
    if (row_floor['floor'] < row_floor['floors_total'])&(row_floor['floor'] != 1):
        return 'другой'
# расчет категории этажа квартиры
df['floor_cat'] = df.apply(floor_categorization, axis=1)
```

Расчет соотношения жилой и общей площади, а также отношения площади кухни к общей

```python
# соотношение жилой и общей площади
df['living_to_total_area'] = (df['living_area'] / df['total_area'])
# соотношение жилой и общей площади
df['kitchen_to_total_area'] = (df['kitchen_area'] / df['total_area'])
```

```python
df.head()
```

**Вывод**
В таблицу добавлены столбцы:
- qm_price - цена квадратного метра в рублях
- weekday, month, year - номера дней недели, месяца и года размещения объявления
- floor_cat - категория этажа квартиры (первый, последний, другой)
- living_to_total_area - доля жилой площади в квартире
- kitchen_to_total_area - доля площади кухни в квартире
